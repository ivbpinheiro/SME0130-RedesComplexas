```python
import networkx as nx
import numpy as np
from numpy  import *
import math
```

```python
def degree_distribution(GER):
    vk = dict(GER.degree())
    vk = list(vk.values()) 
    maxk = np.max(vk)
    mink = np.min(min)
    kvalues= arange(0,maxk+1) 
    Pk = np.zeros(maxk+1) 
    for k in vk:
        Pk[k] = Pk[k] + 1
    Pk = Pk/sum(Pk) 
    return kvalues,Pk
```

```python
def momment_of_degree_distribution(G,m):
    k,Pk = degree_distribution(G)
    M = sum((k**m)*Pk)
    return M
```

```python
def average_clustering (self):
  avc = nx.average_clustering(self)
  return avc
```

```python
def average_degree(self):
  N = len(self)
  M = self.number_of_edges()
  return 2*M/N
```

```python
def random_graphs(N, p):
  return nx.gnp_random_graph(N, p)
```

```python
def random_graphs_small(N, av_degree, p):
  return nx.watts_strogatz_graph(N, av_degree, p, seed=None)
```

```python
def distances(self):
    N = len(self)
    vl = []
    for i in np.arange(0, N):
        for j in np.arange(i+1, N):
            if(i != j):
                aux = nx.shortest_path(self,i,j)
                vl.append(len(aux)-1)
    return vl

def shannon_entropy(G):
    k,Pk = degree_distribution(G)
    H = 0
    for p in Pk:
        if(p > 0):
            H = H - p*math.log(p, 2)
    return H
```

**Exercicio 1:**

Gere um grafo aleatório com N = 1000 e p = 0.1. Qual o valor do grau médio, segundo momento do grau e coeficiente de aglomeração médio (average clustering coefficient)? Considere os valores mais próximos.

```python
GER = random_graphs(1000, 0.1)

print("Average degree %.3f"%average_degree(GER))

print("Second momment of degree %.3f"%momment_of_degree_distribution(GER,2))

print("Average clustering coefficient %.3f"%average_clustering(GER))    
```

**Exercicio 3:**

Considere uma rede aleatória (Erdos-Renyi) com N=1000 vértices. Qual o valor da entropia de Shannon do grau para ⟨k⟩=5, ⟨k⟩=10, ⟨k⟩=50? Aproxime os valores.

```python
N = 1000
for av_degree in [5, 10, 50]:
  p = av_degree/(N-1)
  G = random_graphs(N, p)
  H = shannon_entropy(G)

  print("For average K equals", av_degree, "Shannon entropy is %.1f"%H)
```

**Exercicio 5:**

Considere o modelo de Erdos-Renyi. Gere redes com grau médio igual a 5, 10 e 50 e N=1000. Qual o valor da assortatividade? Aproxime os valores.

```python
N = 1000
for av_degree in [5, 10, 50]:
  p = av_degree/(N-1)
  G = random_graphs(N, p)
  r = nx.degree_assortativity_coefficient(G)
  print("Assortativity for k", av_degree, "is %.1f"%r)
```
