```python
def momment_of_degree(self, m):  
  return np.sum([(self.degree(i)**m)/len(self) for i in self.nodes]);
```

**Exercicio 2:**

Implemente uma rotina para calcular a medida de complexidade. Qual o valor da complexidade para a rede de energia elétrica dos EUA (Base powergrid).

```python
def mean_degree(self):
  print(self.degree())
  return np.sum([degree/len(self) for _, degree in self.degree()]); 
```

```python
def complexity_coefficient(self):
  return "%3.4f"%(momment_of_degree(self, 2) / mean_degree(self));
```

```python
ex2 = read_file('powergrid');
print(f'O Coeficiente de complexidade da rede é: {complexity_coefficient(ex2)}');
```

**Exercicio 3:**

Implemente uma rotina para calcular a entropia de Shannon e calcule essa medida para a base de estradas da Europa (base euroroad).

```python
def degree_distribution(G):
    vk = dict(G.degree())
    vk = list(vk.values())  # we get only the degree values
    vk = np.array(vk)
    maxk = np.max(vk)
    mink = np.min(vk)
    kvalues= np.arange(0,maxk+1) # possible values of k
    Pk = np.zeros(maxk+1) # P(k)
    for k in vk:
        Pk[k] = Pk[k] + 1
    Pk = Pk/sum(Pk) # the sum of the elements of P(k) must to be equal to one
    return kvalues,Pk
```

```python
def shannon_entropy(G):
    k,Pk = degree_distribution(G)
    H = 0
    for p in Pk:
        if(p > 0):
            H = H - p*math.log(p, 2)
    return "%3.4f"%H
```

```python
ex3 = read_file('euroroad');
print(f'A Entropia de Shannon é: { shannon_entropy(ex3)}');
```

**Exercicio 4:**

Calcule as medidas transitivity e average clustering coefficient para a base de dados de aeroportos do EUA (base usairport.txt).

```python
def transitivity (self):
  CC = (nx.transitivity(self)) 
  return "%3.4f"%CC
```

```python
def average_clustering (self):
  avc = nx.average_clustering(self)
  return "%3.4f"%avc
```

```python
ex4 = read_file('usairport');

print(f'A Transitividade é: {transitivity(ex4)}')
print(f'O Agrupamento Media é: {average_clustering(ex4)}')
```
